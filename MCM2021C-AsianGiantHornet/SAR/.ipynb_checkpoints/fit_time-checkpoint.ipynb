```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

```python
def func(x, x0, c0, c1, c2):
    return max(0.0, c1 * np.sin(c0 * x + x0) + c2)
# -526.23441896 * sin(0.53359878 * (x / 365.0 * 12 + 1) + 7.06906783) + 374.69554147
```

```python
f = open("time4.txt", "r")
cnt = f.read().split('\n')[0:12]
for i in range(0, 12):
    cnt[i] = int(cnt[i])
cnt = cnt / np.sum(cnt)

x = np.linspace(0, 11, 12)
dx = np.linspace(0, 12, 24)
ix = np.linspace(1, 12, 12)
# plt.plot(cnt)
# plt.show()
```

```python
popt, pcov = curve_fit(func, ix, cnt, bounds = ((0, np.pi / 6 - 0.01, -10000, -10000), (12, np.pi / 6 + 0.01, 10000, 10000)))
print(popt)
print(pcov)
```
